```python
features = ['country', 'genre', 'year_added', 'month_added']
music2 = music[music.plays <200]
X = music2[features]
X = sm.add_constant(X)
y = music2[["plays"]]
```

```python
X = pd.get_dummies(X, columns = ['country', 'genre', 'year_added', 'month_added'])
```

```python
X_train, X_test , y_train, y_test = train_test_split(X,y, test_size =0.25, random_state =246)
print(X_train.shape, y_train.shape)
print(X_test.shape, y_test.shape)
```

```python
from sklearn.linear_model import LinearRegression
rModel = linear_model.LinearRegression(normalize = True)
```

```python
rModel = sm.OLS(y_train, X_train)


# Fit the model
result = rModel.fit()
```

```python
result.summary()
```

```python
y_pred = result.predict(X_test)
plt.scatter(y_test, y_pred)

# Add a line for perfect correlation
plt.plot([x for x in range(0,200)],[x for x in range(0,200)], color='g')

# Label it nicely
plt.xlabel('Actual Plays')
plt.ylabel('Predicted Plays')
_ = plt.xlim([0,200])
_ = plt.ylim([0,200])
plt.show()
```

```python
def rmse(predictions, targets):
    return np.sqrt(((predictions - targets)**2).mean())

def mse(predictions, targets):
    return abs((predictions - targets)).mean()


matches = pd.DataFrame(y_test)
matches.rename(columns = {'plays':'actual'}, inplace=True)
matches["predicted"] = y_pred

print('RMSE: ' + str(rmse(matches["actual"], matches["predicted"])))
print('MSE: ' + str(mse(matches["actual"], matches["predicted"])))
```
